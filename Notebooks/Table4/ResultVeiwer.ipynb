```python
import pandas as pd

profiles = pd.read_csv('tadarimalstmAE.csv')
profiles
```

```python
tadgan=profiles[profiles["pipline"] =='tadgan'].reset_index(drop=True)
arima=profiles[profiles["pipline"] =='arima'].reset_index(drop=True)
lstm_autoencoder=profiles[profiles["pipline"] =='lstm_autoencoder'].reset_index(drop=True)
lstm_autoencoder                                                                                      
```

```python
arima_MSL= pd.read_csv('ARIMA_MSL.csv')
arima_SMAP= pd.read_csv('ARIMA_SMAP.csv')
arima_MSL=pd.read_csv('ARIMA_MSL.csv', index_col=[0])
arima_SMAP=pd.read_csv('ARIMA_SMAP.csv', index_col=[0])

arima_SMAP['prediction_time']=arima_SMAP['Prediction_time']
arima_MSL['prediction_time']=arima_MSL['Prediction_time']
arima_MSL=arima_MSL.drop('Prediction_time',axis=1)
arima_SMAP=arima_SMAP.drop('Prediction_time',axis=1)
arima_SMAP
```

```python

n_epoch_tadgan=150
n_epoch_arima=150
n_epoch_lstm_autoencoder=150

tadgan['fit_time']=tadgan['fit_time']/n_epoch_tadgan
tadgan['prediction_time']=tadgan['prediction_time']/n_epoch_tadgan
#arima['value']=arima['value']/(n_epoch_arima)
lstm_autoencoder['fit_time']=lstm_autoencoder['fit_time']/n_epoch_lstm_autoencoder
lstm_autoencoder['prediction_time']=lstm_autoencoder['prediction_time']/n_epoch_lstm_autoencoder

profiles= pd.concat([tadgan,arima_MSL,arima_SMAP,lstm_autoencoder])
profiles
```
